```python
# Combine features and target for easy correlation checking
df_corr_lat = df_cleaned.copy()
df_corr_lat['target_latitude'] = df_corr_lat['Latitude']
df_corr_lat['target_longitude'] = df_corr_lat['Longitude']

# Compute correlations with targets
corr_lat = df_corr_lat.corr(numeric_only=True)['target_latitude'].sort_values(ascending=False)
corr_lon = df_corr_lat.corr(numeric_only=True)['target_longitude'].sort_values(ascending=False)

print("Top correlations with Latitude:\n", corr_lat.head(10))
print("\nTop correlations with Longitude:\n", corr_lon.head(10))
```

```python
# Define targets
y_lat = df_cleaned['Latitude']
y_lon = df_cleaned['Longitude']

# Define features
X = df_cleaned.drop(columns=['Latitude', 'Longitude', 'Date','Updated On'])
```

```python
# Drop non-numeric / string columns that shouldn't be features
X = X.drop(columns=['Block', 'Case Number', 'IUCR','Description','FBI Code','Location'], errors='ignore')
X = X.drop(columns=['X Coordinate', 'Y Coordinate','ID'], errors='ignore')
```

```python
from sklearn.model_selection import train_test_split

# Split the feature set and both targets
X_train, X_test, y_lat_train, y_lat_test = train_test_split(X, y_lat, test_size=0.2, random_state=42)
_, _, y_lon_train, y_lon_test = train_test_split(X, y_lon, test_size=0.2, random_state=42)
```

```python
# Show all object (string) columns in your feature set
print(X.dtypes[X.dtypes == 'object'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Train model for latitude
lat_model = RandomForestRegressor(n_estimators=30, max_depth=15, random_state=42)

lat_model.fit(X_train, y_lat_train)

# Predict latitude
y_lat_pred = lat_model.predict(X_test)

```

```python
# Predict on the test set
y_lat_pred = lat_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Evaluation metrics
lat_mae = mean_absolute_error(y_lat_test, y_lat_pred)
lat_rmse = np.sqrt(mean_squared_error(y_lat_test, y_lat_pred))

print(f"Latitude Model - MAE: {lat_mae:.6f}")
print(f"Latitude Model - RMSE: {lat_rmse:.6f}")
```

```python
# Train longitude model
lon_model = RandomForestRegressor(n_estimators=30, max_depth=15, random_state=42)
lon_model.fit(X_train, y_lon_train)

# Predict and evaluate
y_lon_pred = lon_model.predict(X_test)

```

```python
lon_mae = mean_absolute_error(y_lon_test, y_lon_pred)
lon_rmse = np.sqrt(mean_squared_error(y_lon_test, y_lon_pred))

print(f"Longitude Model - MAE: {lon_mae:.6f}")
print(f"Longitude Model - RMSE: {lon_rmse:.6f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_lon_test, y_lat_test, color='blue', label='True Locations', alpha=0.4)
plt.scatter(y_lon_pred, y_lat_pred, color='green', label='Predicted Locations', alpha=0.4)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Predicted vs True Crime Locations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Feature importance for latitude model
importance = lat_model.feature_importances_
feat_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importance})
feat_df = feat_df.sort_values('Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_df.head(15))
plt.title("Top 15 Important Features (Latitude Prediction)")
plt.tight_layout()
plt.show()
```

```python
import folium

# Center map on Chicago
crime_map = folium.Map(location=[41.8781, -87.6298], zoom_start=11)

# Add true crime locations (blue)
for lat, lon in zip(y_lat_test[:100], y_lon_test[:100]):
    folium.CircleMarker(location=[lat, lon], radius=3, color='blue', fill=True).add_to(crime_map)

# Add predicted crime locations (red)
for lat, lon in zip(y_lat_pred[:100], y_lon_pred[:100]):
    folium.CircleMarker(location=[lat, lon], radius=3, color='red', fill=True).add_to(crime_map)

crime_map.save("crime_predictions_map.html")
```
